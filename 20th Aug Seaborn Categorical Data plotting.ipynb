```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.catplot(x='day', y = 'total_bill', data=tips, hue='day')
plt.show()
```

```python
sns.swarmplot(x='day', y='total_bill',data=tips, hue='sex')
plt.show()
```

```python
sns.countplot(x='day',data=tips)
plt.yticks(range(10,100,10))
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.grid()
sns.countplot(x='day', hue='time', data=tips)
plt.yticks(range(0,100,10))
plt.show()
```

```python
sns.stripplot(x='size',y='total_bill', data=tips, color='g')
plt.show()
```

```python
sns.stripplot(x='smoker', y='total_bill', data=tips, hue = 'sex')
plt.show()
```
