```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

from pandas.plotting import scatter_matrix
```

```python
#Load data from seaborn
mileage = sns.load_dataset('mpg')
mileage.head()
```

```python
#with two columns
plt.scatter(x=mileage['cylinders'],y=mileage['mpg'])
plt.xlabel['Cyl']
plt.ylabel['mpg']
plt.show()
```

```python
#Lets plot scatter matrix to check relationship of all columns
```

```python
#Scatter_matrix
scatter_matrix(mileage, figsize = (15,10), color = 'b')
plt.show
```
