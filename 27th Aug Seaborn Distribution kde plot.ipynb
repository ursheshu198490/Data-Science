Visualizing Distribution of the Data

a) distplot()
b) kdeplot()
c) jointplot()

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from numpy.random import randn

import warnings
warnings.filterwarnings('ignore')
```

```python
x = randn(10000)
```

```python
sns.distplot(x,kde=True)
plt.show()
```

```python
sns.kdeplot(x)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.tail()
```

```python
x = tips['total_bill']
y = tips['tip']
```

```python
sns.jointplot(x=x, y=y)
plt.show()
```
