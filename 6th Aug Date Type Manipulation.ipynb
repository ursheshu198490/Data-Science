```python
import numpy as np

import pandas as pd
```

```python
date_range = pd.date_range('4/12/2021',periods=12,freq='D')
dates = pd.DataFrame(date_range,columns = ['Date'])
```

```python
dates
```

```python
dates['Day']=dates['Date'].dt.day
dates['Month']=dates['Date'].dt.month
dates['Year']=dates['Date'].dt.year

dates
```

```python
df = pd.DataFrame({'DATE_TIME':['Jan 1 12:00 am 2021','Jan 2 12:00 am 2022']})
df
```

```python
df['Date']=pd.to_datetime(df['DATE_TIME'],format = '%b %d %I:%M %p %Y')
df
```

```python
df['Day']=df['Date'].dt.day
df['Month']=df['Date'].dt.month
df['Year']=df['Date'].dt.year

df
```
