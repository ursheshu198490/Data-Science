```python
#Replace wrong name with correct name by using replace method
df['Country'].replace({'UAS':'USA'},inplace=True)
df
```

```python
df['Country'].fillna(value=df['Country'].mode()[0],inplace=True)
df
```

```python
#Lets try adding both income and keep total in new column called Salary
df['Total Income'] = df['Income.1'] + df['Income.2']
```

```python
#As the Income 1 Contains special character dollar sign so it is treated as object data and you cannot add it to int. 
#We will try to fix dollar[$] sign.
#To fix the issue we have to remove the sign and convert it into float data type
```

```python
#Replace  sign with Empty
df['Income.1'] = df['Income.1'].str.replace("$","")
df
```

```python
#Replace space with nulls so that we can treat all nulls together
df['Income.1'] = df['Income.1'].replace("",np.nan)
df
```

```python
df.info()
```

```python
#Since Income.1 had '$' sign in it and still will be considered as string. Lets convert into float and convert all Nan's 
#later

df['Income.1'] = df['Income.1'].astype(float)
```

```python
df.info()
```

```python
df['Income.1'].fillna(df.groupby('gender')['Income.1'].transform("mean"),inplace=True)
df
```

```python
df.info()
```

```python
#Create a new column to store both the incomes
df['Total Income'] = np.ceil(df['Income.1'] + df['Income.2'])
df
```

```python
# Split full name into first and second
df['Employee Full Name'].str.split(" ",expand = True)
```

```python
df['First_name'] = df['Employee Full Name'].str.split(" ").str.get(0)
df['Second_name'] = df['Employee Full Name'].str.split(" ").str.get(1)
df
```

```python
#Move columns First and Last name to the First and second position.
```

```python
column_to_move = df['First_name']
df = df.drop('First_name', axis=1)
df.insert(1, 'First_name', column_to_move)

df
```

```python
column_to_move = df['Second_name']
df = df.drop('Second_name', axis=1)
df.insert(2, 'Second_name', column_to_move)

df
```

```python
df.drop('Employee Full Name', axis = True, inplace = True)
```
